```python
data = pd.read_table('./data/GM12878_primary_interchromosomal/100kb_resolution_interchromosomal/chr1_chr2/MAPQGE30/chr1_2_100kb.RAWobserved', header=None)
```

```python
counts = data[2].value_counts().to_dict()
count_data = pd.DataFrame(list(counts.items()), columns = ['value', 'frequency'])
sorted_count_data = count_data.sort_values(by=['frequency'])
sort_data = sorted_count_data.sort_values(by=['value'], ascending=False)
sorted_count_data
number_of_contacts = len(data)
one_percent = round(number_of_contacts/100)
one_percent
#frequences.columns = ["value", "frequency"]

#plt.hist(count_data['frequency'].tail(30))
```

```python
sort_data
freq_sum = 0
data_matrix = np.matrix(sort_data)

def find_cutoff(dataset, percent):
    
    ### dataset = sorted dataset with value and frequency data, result of count_values ###
    ### percent = 1 = 1%, 5 = 5% and etc ###
    
    data_matrix = np.matrix(dataset)
    number_of_contacts = len(dataset)
    percentage = round(number_of_contacts*(percent/100))
    freq_sum = 0
    
    for element in data_matrix:
        freq_sum += element[0,1]
        if freq_sum >= one_percent:
            return element[0,0]
```

```python
find_cutoff(sort_data, 1)
```
